```python
%matplotlib inline
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
d = path.dirname('__file__')

# Read the whole text.
text = open(path.join(d, 'search_terms.txt')).read()
```

```python
# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open(path.join(d, "brand_logo_green.png")))
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=STOPWORDS.add("said"))
# generate word cloud
wc.generate(text)

# store to file
#wc.to_file(path.join(d, "search_popularity.png"))
```

```python
# show
plt.imshow(wc)
plt.axis("off")
#plt.figure()
#plt.imshow(alice_mask, cmap=plt.cm.gray)
#plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud().generate(text)

#plt.imshow(wordcloud)
#plt.axis("off")

# take relative word frequencies into account, lower max_font_size
wordcloud = WordCloud(relative_scaling=.5, width=800, height=400).generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file(path.join(d, "search_popularity.png"))
```
